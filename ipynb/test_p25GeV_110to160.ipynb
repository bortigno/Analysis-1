# Setup.

- Mass Range [110, 160] 0.25Gev per point
- Single Gaus Singal Model
- 01JetsTightBB Category

# BWZRedux. r-scan. Use Asimov Dataset

```
combine -M MultiDimFit -t -1 --algo=grid --points=200 --rMin=0 --rMax=5 --setPhysicsModelParameters pdfindex_cat6=0,MH=125 --freezeNuisances pdfindex_cat6,MH --expectSignal=1 datacard__cat6__SingleGaus.root -m 125 -S 0 --cminDefaultMinimizerType=Minuit2 --saveWorkspace -n BWZRedux
```

````python
canvas = R.TCanvas("Some Canvas", "Some CAnvas", 800, 600)
pathToDir = "/afs/cern.ch/work/v/vkhriste/public/Higgs2Mu/combine_tests/test_p25GeV_110to160"
f = R.TFile(os.path.join(pathToDir, "higgsCombineBWZRedux.MultiDimFit.mH125.root"))
t = f.Get("limit")
t.Draw("2*deltaNLL:r","","*")
canvas.Draw()
````
