### Running r-scan for cat6 (01JetsTightBB) on Asimov Dataset w/o freezing any model parameters. pdfindex is on BWZRedux

combine -M MultiDimFit -t -1 --algo=grid --points=200 --rMin=0 --rMax=5 --setPhysicsModelParameters pdfindex_cat6=0,MH=125 --freezeNuisances pdfindex_cat6,MH --expectSignal=1 datacard\_\_cat6\_\_SingleGaus.root -m 125 -S 0 --cminDefaultMinimizerType=Minuit2  --saveWorkspace
